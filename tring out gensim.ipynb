# Convert pandas df row to w2v / glove input

```python
'''
I thnks things like `` sorta ``e removed or tf-idf or chi squared would take care
'''    

from nltk.tokenize import sent_tokenize, word_tokenize
lst = []
for row in df_preprocessed['review']:
    data =  word_tokenize(row)
    lst.append(data)
    #lst.append(data)
    
```

```python
print(lst[5:10])
```

```python
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
from matplotlib import pyplot
%matplotlib inline
```

```python
%%time
model = Word2Vec(lst, min_count=1,)
'''
Set values for various parameters
num_features = 300    # Word vector dimensionality                      
min_word_count = 40   # Minimum word count                        
num_workers = 4       # Number of threads to run in parallel
context = 10          # Context window size                                                                                    
downsampling = 1e-3   # Downsample setting for frequent words
'''
```

#  Load locally trained w2v

```python
w2Vec_Embedding = {} #dict()
                                                               
f = open('./Dataset/imdb-vectors-2000.txt', encoding="utf8")  #100d

for line in f:
	values = line.split() # split by lines
	word = values[0]# get 1st word in the line i.e the "word" itself
	coefs = np.asarray(values[1:], dtype='float32') # the preceding vector is the vector itself
	w2Vec_Embedding[word] = coefs# its a dict of word(key) and vector(value)
f.close()
```
